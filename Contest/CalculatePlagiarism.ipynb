```python
from sctokenizer import Source
import itertools
import re

def tokens(file, language):
    src = Source.from_file(file, lang=language)
    tokens = src.tokenize()
    tokens_list = []
    tokens_list_final = []
    for token in tokens:
        tokens_list.append(str(token)[(str(token).find(',')):(str(token).rfind(','))])
    for i in range(0, len(tokens_list)):
        tokens_list[i] = re.sub('[^a-zA-Z]', '', tokens_list[i])
        tokens_list_final.append(tokens_list[i].replace('TokenType', ''))
    #for i in range(0, len(tokens_list_final)):
        #print(tokens_list_final[i])
    unique_tokens = []
    unique = set(tokens_list_final)
    for token in unique:
        unique_tokens.append(token)
    #for i in range(0, len(unique_tokens)):
        #print(unique_tokens[i])
    for i in range(0, len(tokens_list_final)):
        for j in range(0, len(unique_tokens)):
            if tokens_list_final[i] == unique_tokens[j]:
                tokens_list_final[i] = j + 1
    res_str = ''
    for i in range(0, len(tokens_list_final)):
        res_str = res_str + str(tokens_list_final[i])
    #print(res_str)
    n_gramms = []
    n_gramms_final = []
    n_gramms_unique = []
    for i in range(0, len(res_str)):
        n_gramms.append(res_str[i:i+4])
    for j in range(0, len(n_gramms)):
        if(len(n_gramms[j]) == 4):
            n_gramms_final.append(n_gramms[j])
    #for i in range(0, len(n_gramms_final)):
        #print(n_gramms_final[i])
    unique_n_gramms = set(n_gramms_final)
    for n_gramm in unique_n_gramms:
        n_gramms_unique.append(n_gramm)
    return n_gramms_unique
```
