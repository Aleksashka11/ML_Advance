```python
from sklearn.tree import DecisionTreeClassifier
from collections import Counter
```

```python
def bootstrap_sample(X, y):
    n_samples = X.shape[0]
    idxs = np.random.choice(n_samples, n_samples, replace=True)
    bootstrap_sample_x = pd.DataFrame()
    bootstrap_sample_y = pd.DataFrame()
    bootstrap_sample_x = X.iloc[idxs]
    bootstrap_sample_y = y.iloc[idxs]
    return bootstrap_sample_x, bootstrap_sample_y

def most_common_label(y):
    counter = Counter(y)
    most_common = counter.most_common(1)[0][0]
    return most_common
```

```python
class RandomForestClassifier:
    def __init__(self, n_estimators=100, criterion='gini',
                 max_depth=None, min_samples_split=2,
                 min_samples_leaf=1, min_weight_fraction_leaf=0.0,
                 max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0,
                 min_impurity_split=None, bootstrap=True, random_state=None, max_samples=None):
        self.n_estimators = n_estimators
        self.criterion = criterion
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.min_weight_fraction_leaf = min_weight_fraction_leaf
        self.max_features = max_features
        self.max_leaf_nodes = max_leaf_nodes
        self.min_impurity_decrease = min_impurity_decrease
        self.min_impurity_split = min_impurity_split
        self.bootstrap = bootstrap
        self.random_state = random_state
        self.max_samples = max_samples

        self.estimators = []
        
    def fit(self, X, y):
        self.estimators = []
        for i in range(self.n_estimators):
            tree = DecisionTreeClassifier(criterion=self.criterion, max_depth=self.max_depth, 
                                          min_samples_split=self.min_samples_split, min_samples_leaf=self.min_samples_leaf,
                                          min_weight_fraction_leaf=self.min_weight_fraction_leaf, max_features=self.max_features,
                                          max_leaf_nodes=self.max_leaf_nodes, min_impurity_decrease=self.min_impurity_decrease, 
                                          random_state=self.random_state)
        X_samp, y_samp = bootstrap_sample(X, y)
        tree.fit(X_samp, y_samp)
        self.estimators.append(tree)
            
    def predict(self, X):
        tree_preds = np.array([tree.predict(X) for tree in self.estimators])
        tree_preds = np.swapaxes(tree_preds, 0, 1)
        y_pred = [most_common_label(tree_pred) for tree_pred in tree_preds]
        return np.array(y_pred)
```

```python
rfc = RandomForestClassifier(n_estimators=5, bootstrap=True, max_samples=1000)
```

```python
rfc.fit(X_train, y_train)
```

```python
def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred)/len(y_true)
    return accuracy
```

```python
y_pred = rfc.predict(X_test)
accuracy(y_test, y_pred)
```
